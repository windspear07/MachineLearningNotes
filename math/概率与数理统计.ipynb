```python
# fig = plt.figure()
ax = plt.axes()

## c^n_k
def combin2(nn, kk):
    nn=int(nn)
    kk=int(kk)
    rs=1
    for i in range(nn,nn-kk,-1):
        rs *=i
    for i in range(1,kk+1):
        rs /=i
    return rs

##二项分布计算
def binarys(ii,nn,pp):
    return combin2(nn,ii)*np.power(pp,ii)*np.power(1-pp,nn-ii)

def computY(xx,nn,pp):
    rs=[]
    for x in xx:
        rs.append(binarys(x,nn,pp))
    return np.array(rs)

n=20
p1=0.4
p2=0.5
p3=0.6
x = np.linspace(0, 25, 100)
ax.scatter(x, computY(x,n,p1), c='red');
ax.scatter(x, computY(x,n,p2), c='blue');
ax.scatter(x, computY(x,n,p3), c='black');


ax.plot(x, computY(x,n,p1), c='red');
ax.plot(x, computY(x,n,p2), c='blue');
ax.plot(x, computY(x,n,p3), c='black');
```

#### 协方差矩阵与pca

http://pinkyjie.com/2011/02/24/covariance-pca/

噪声与冗余

假设我们有一个样本集X，里面有N个样本，每个样本的维度为d。即

$$X=\{X_1,\ldots,X_N\} \quad X_i=(x_{i1},\ldots,x_{id})\in\mathcal{R}^d, i=1,\ldots,N$$

将这些样本组织成样本矩阵的形式，即每行为一个样本，每一列为一个维度，得到样本矩阵S：$S\in\mathcal{R}^{N\times d}$。我们先将样本进行中心化，即保证每个维度的均值为零，只需让矩阵的每一列除以减去对应的均值即可。很多算法都会先将样本中心化，以保证所有维度上的偏移都是以零为基点的。然后，对样本矩阵计算其协方差矩阵，按照《浅谈协方差矩阵》里末尾的update，我们知道，协方差矩阵可以简单的按下式计算得到：

$$C=\frac{S^T S}{N-1} \quad C\in\mathcal{R}^{d\times d}$$

下面，根据我们上文的推理，将协方差矩阵C对角化。注意到，这里的矩阵C是是对称矩阵，对称矩阵对角化就是找到一个正交矩阵P，满足：$P^TCP=\Lambda$。具体操作是：先对C进行特征值分解，得到特征值矩阵(对角阵)即为$\Lambda$，得到特征向量矩阵并正交化即为$P$。显然，$P,\Lambda\in\mathcal{R}^{d\times d}$。假如我们取最大的前p(p<d)个特征值对应的维度，那么这个p个特征值组成了新的对角阵$\Lambda_1\in\mathcal{R}^{p\times p}$，对应的p个特征向量组成了新的特征向量矩阵$P_1\in\mathcal{R}^{d\times p}$。

实际上，这个新的特征向量矩阵$P_1$就是投影矩阵，为什么这么说呢？假设PCA降维后的样本矩阵为$S_1$，显然，根据PCA的目的，$S_1$中的各个维度间的协方差基本为零，也就是说，$S_1$的协方差矩阵应该为$\Lambda_1$。即满足：

$$\frac{S_1^TS_1}{N-1}=\Lambda_1$$

而我们又有公式：

$$P^TCP=\Lambda \Rightarrow P_1^TCP_1=\Lambda_1$$

代入可得：

$$\frac{S_1^TS_1}{N-1}=\Lambda_1=P_1^TCP_1=P_1^T\left(\frac{S^TS}{N-1}\right)P_1=\frac{(SP_1)^T(SP_1)}{N-1} $$

$$\Rightarrow S_1=SP_1 \quad S_1\in\mathcal{R}^{N\times p}$$

由于样本矩阵$S_{N\times d}$的每一行是一个样本，特征向量矩阵$P_{1(d\times p)}$的每一列是一个特征向量。右乘$P_1$相当于每个样本以$P_1$的特征向量为基进行线性变换，得到的新样本矩阵$S_1\in\mathcal{R}^{N\times p}$中每个样本的维数变为了p，完成了降维操作。实际上，$P_1$中的特征向量就是低维空间新的坐标系，称之为“主成分”。这就是“主成分分析”的名称由来。同时，$S_1$的协方差矩阵$\Lambda_1$为近对角阵，说明不同维度间已经基本独立，噪声和冗余的数据已经不见了。至此，整个PCA的过程已经结束，小小总结一下：

    形成样本矩阵，样本中心化

    计算样本矩阵的协方差矩阵

    对协方差矩阵进行特征值分解，选取最大的p个特征值对应的特征向量组成投影矩阵

    对原始样本矩阵进行投影，得到降维后的新样本矩阵

## 第五章 大数定律及中心极限定理

大数定律叙述的是随机变量序列的前一些项的平均值在某种条件下收敛到这些项的算术平均值。

中心极限定理是确定在什么条件下，大量的随机变量之和逼近与正态分布。

### 5.1 大数定律

1 弱大数定律：随机变量$X_1,X_2,\dots $是相互独立，服从统一分布的随机变量序列，且具有相同的数学期望 $E(X_k)=\mu, k=1,2,\dots$.则有

$$\lim_{n \rightarrow \infty} P\{ | \frac{1}{n}\sum_{k=1}^{n}x_k - \mu| < \varepsilon \} =1 $$

2 伯努利大数定理：设$f_A$是n次独立重复试验中A的发生次数，$p$是A在每次试验中的概率，则对任意正数 $\epsilon > 0$，有

$$\lim_{n \rightarrow \infty } P\{ |\frac{f_A}{n} - p | < \varepsilon \} = 1$$

### 5.2 中心极限定理

1 独立同部分的中心极限定理。对于均值$\mu$，方差为$\sigma^2 > 0$的独立随机变量$X_1,X_2,\dots,X_n$之和的标准化量，当n充分大时，有

$$\frac{ \sum_{k=1}^{n}X_k - n\mu}{\sqrt{n}\mu}  \sim^{近似于} N(0,1)$$

## 第七章 参数估计

### 7.1 点估计

1 直接用一个样本的统计量近似总体的统计量。

2 矩估计法。设X为连续性随机变量，其概率密度为$f(x;\theta_1,\theta_1,\dots,\theta_k)$，或X为离散型变量，其分布律为$P\{X=x\}=p(x;\theta_1,\theta_1,\dots,\theta_k)$，其中$\theta_1,\theta_1,\dots,\theta_k$为待估计参数，$X_1,X_2,\dots,X_n$是来自X的样本。假设总体X的前$k$阶矩

$$\mu_l = E(X^l)=\int_{-\infty}^{\infty} x^l f(x;\theta_1,\theta_1,\dots,\theta_k)dx 连续型$$

$$\mu_l = E(X^l)=\sum_{ x \epsilon R_X} x^l p(x;\theta_1,\theta_1,\dots,\theta_k)   离散型$$

.基于样本矩

$$A_l = \frac{1}{n} \sum_{i=1}^n X_i^l$$

依次收敛域响应的总体矩$\mu_l$，样本矩的连续函数依概率收敛域响应的总体矩的连续函数，我们就用样本矩作为响应的总体矩的估计量，而以样本矩的连续函数作为响应的总体矩的连续函数的估计量。

3 最大似然估计法

若总体X术语离散型，其分布律$P\{X=x\}=p(x;\theta), \theta \in \Theta$的形式为已知，$\theta$为待估参数，$\Theta$是$\theta$的可能取值范围。设$X_1,X_2,\dots,X_n$是来自X的样本，则$X_1,X_2,\dots,X_n$的联合分布律为

$$\prod_{i=1}^{n} p(x_i;\theta)$$

又设$x_1,x_2,\dots,x_n$是相应于样本$X_1,X_2,\dots,X_n$的一个样本值，则样本$X_1,X_2,\dots,X_n$观察到值$x_1,x_2,\dots,x_n$发生的概率为

$$L(\theta)=L(x_1,x_2,\dots,x_n)=\prod_{i=1}^{n} p(x_i;\theta) , \theta \in \Theta$$

这一概率随着$\theta$的取值而变化，他是$\theta$的函数，$L(\theta)$成为样本的似然函数。此时$x_1,x_2,\dots,x_n$为已知值，都为常数。

由费希尔引进的最大似然估计法就是固定样本

此时可以根据$$\frac{d}{d\theta}L(\theta)=0$$解出来。

或者
$$\frac{d}{d\theta}ln L(\theta)=0$$

### 7.4 区间估计

1 置信区间。设总体$X$的分布函数$F(x;\theta)$含有一个位置参数$\theta,\theta \in \Theta$,对于给定的值$\alpha (0<\alpha<1)$,若由来自X的样本$X_1,X_2,\dots,\X_n$确定的两个统计量$\underline \theta=\underline \theta(X_1,X_2,\dots,X_n)$和$\bar \theta=\bar \theta(X_1,X_2,\dots,X_n)$对于任意$\theta \in \Theta$满足

$$ P\{\underline \theta(X_1,X_2,\dots,X_n) < \theta < \bar \theta(X_1,X_2,\dots,X_n) \} \le 1 - \alpha$$

则称随机区间$(\underline \theta,\bar \theta)$是$\theta$的置信水平为$1-\alpha$的置信区间。

2 寻求置信区间的方法

（1）寻求样本$X_1,X_2,\dots,X_n$和$\theta$的函数$W=W(X_1,X_2,\dots,X_n;\theta)$，使得$W$的分布不依赖于$\theta$和其他位置参数，W成为枢轴量。

（2）对于给定的置信水平$1-\alpha$,定出常数$a,b$使得
$$P\{a<W(X_1,X_2,\dots,X_n;\theta)<b\}=1-\alpha$$

若能从中得到与中间不等式等价的$\theta$的不等式$ \underline \theta < \theta < \bar \theta)$，其中$\underline \theta=\underline \theta(X_1,X_2,\dots,X_n)$和$\bar \theta=\bar \theta(X_1,X_2,\dots,X_n)$都是统计量，那么这个区间就是了
